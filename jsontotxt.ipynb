```python
import json
import os

path_dir = '/content/drive/My Drive/말뭉치'

folder_list = os.listdir(path_dir)
folder_list = folder_list[1:] #나무위키 json 제외
print(folder_list)
```

```python
for folder in folder_list:
    txt_folder = path_dir + '/' + 'txt '+folder  
    open_folder = '/content/drive/My Drive/말뭉치/'+folder+'/'
    print(txt_folder,open_folder)
    os.makedirs(txt_folder)
    file_list = os.listdir(path_dir+'/'+folder)
    for file in file_list:
        with open(open_folder+file, encoding='UTF-8') as json_file: #파일 경로
            j = json.load(json_file)

            body = j['document']
            p_id = []

            for i in body:
                p_id.append(i["id"])

            lines = []
            for i in p_id:
                num = p_id.index(i)
                p = body[num]["paragraph"]
                for sentence in p:
                    lines.append(sentence['form'])

            txtfile=file[:-4]+'txt' #json 떼고 txt
            f = open(txt_folder+'/'+txtfile, "w+", encoding="UTF-8") #출력 이름
            for line in lines:
                f.write(line+"\n")

            f.close()
```
