```python
from torch.utils.data.dataset import Dataset
from transformers import PreTrainedTokenizer
from transformers import logging
from typing import Dict
import warnings 
from google.colab import drive #Íµ¨Í∏ÄÎìúÎùºÏù¥Î∏å Í≤ΩÎ°ú Í∞ÄÏ†∏Ïò§Í∏∞ ÏúÑÌï¥ÏÑú Ï∂îÍ∞Ä 
# ÎßêÎ≠âÏπò ÎÇ¥Ïùò ÏùºÎ∂Ä ÌäπÏàò Í∏∞Ìò∏ ‚ë†, ‚ñ≤, ‚óÜ, ‚ñ†, ‚óá, ‚ñ£  Ïã†Í≤Ω Ïì∏Íπå? 

drive.mount("/content/gdrive", force_remount=True)

DEPRECATION_WARNING = (
    "This dataset will be removed from the library soon, preprocessing should be handled with the ü§ó Datasets "
    "library. You can have a look at this example script for pointers: {0}"
)

logger = logging.get_logger(__name__)


class LineByLineJsonDataset(Dataset): #Íµ≠Î¶ΩÍµ≠Ïñ¥Ïõê ÏûêÎ£åÎßå ÏùºÎã® Ï†ÅÏö© ÎßêÎ≠âÏπò Í∏∞Í¥ÄÎßàÎã§ json Íµ¨Ï°∞Í∞Ä Îã§Î•º ÏàòÎèÑ ÏûàÏùå 
    """
    This will be superseded by a framework-agnostic approach soon.
    """

    def __init__(self, tokenizer: PreTrainedTokenizer, file_path: str, block_size: int):
        warnings.warn(
            DEPRECATION_WARNING.format(
                "https://github.com/huggingface/transformers/blob/master/examples/pytorch/language-modeling/run_mlm.py"
            ),
            FutureWarning,
        )
        assert os.path.isfile(file_path), f"Input file path {file_path} not found"
        # Here, we do not cache the features, operating under the assumption
        # that we will soon use fast multithreaded tokenizers from the
        # `tokenizers` repo everywhere =)
        logger.info(f"Creating features from dataset file at {file_path}")

        with open(file_path, encoding="utf-8") as f:
            j = json.load(f)

            body = j['document']
            p_id = []

            for i in body:
                p_id.append(i["id"])

            lines = []
            for i in p_id:
                num = p_id.index(i)
                p = body[num]["paragraph"]
                for sentence in p:
                    lines.append(sentence['form'])
        batch_encoding = tokenizer(lines, add_special_tokens=True, truncation=True, max_length=block_size)
        self.examples = batch_encoding["input_ids"]
        self.examples = [{"input_ids": torch.tensor(e, dtype=torch.long)} for e in self.examples]

    def __len__(self):
        return len(self.examples)

    def __getitem__(self, i): #-> Dict[str, torch.tensor]:
        return self.examples[i]

dataset = LineByLineJsonDataset(
    tokenizer = tokenizer,
    file_path = "/content/gdrive/My Drive/Colab Notebooks/NIRW1900000015.json",
    block_size= 128
)
"""
from datasets import load_dataset
dataset = load_dataset('text',data_files={'train':'./datasets/WikiQACorpus/WikiQA-train.txt',
                      'test':'./datasets/WikiQACorpus/WikiQA-test.txt'})

from datasets import load_dataset
dataset = load_dataset('wikicorpus','raw_en')
"""
```

```python
pos_df = pd.read_json(path1,lines=True)
neg_df = pd.read_json(path2,lines=True)
```

```python
pos_df_text = pos_df['text']
neg_df_text = neg_df['text']
```

```python
import regex as re

def basicPreprocess(text):
    try:
        processed_text = text.lower()
        processed_text = re.sub(r'\W +',' ',processed_text)
    except Exception as e:
        print("Exception:",e,",on text:",text)
        return None
    return processed_text
```

```python
type(pos_df_text)
```

```python
pos_df_text = pos_df_text.apply(basicPreprocess).dropna()
neg_df_text = neg_df_text.apply(basicPreprocess).dropna()
```

```python
pos_df_text = pos_df_text.replace('\n'," ")
neg_df_text = neg_df_text.replace('\n'," ")
```
